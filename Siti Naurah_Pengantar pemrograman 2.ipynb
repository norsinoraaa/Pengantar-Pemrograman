```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def my_line(x, m, q):
    y = m * x + q
    return y

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_majors', engine='openpyxl')

my_dataset3 = my_dataset[my_dataset.Epoch == 'three']

x = np.linspace(52.5, 62, 100)
y = my_line(x, m=0.3, q=-10.3)

fig, ax = plt.subplots()

ax.scatter(my_dataset3.SiO2, my_dataset3.K2O, marker='o', color="purple", edgecolor="#cacaca", label=r'$3^{rd}$ Epoch')
ax.plot(x, y, color="#ef3ece")

ax.annotate(r'What is the 1$\sigma$ for this point?', xy=(52.5, 6.99), xytext=(52.3, 9), arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="#ef3ece"))
ax.text(52.4, 5.6, r'$ Na_2O = 0.3 \cdot SiO_2 -10.3$', dict(size=10, rotation=28.4))

ax.text(56, 5.8, r'$ \mu_{SiO_2} = \frac {a_{1}+a_{2}+\cdots +a_{n}}{n}$ = ' + '{:.1f} [wt.%]'.format(57.721),  dict(size=11.5))

ax.set_xlabel(r'SiO$_2$ [wt%]')
ax.set_ylabel(r'K$_2$O [wt%]')

ax.legend()
```
